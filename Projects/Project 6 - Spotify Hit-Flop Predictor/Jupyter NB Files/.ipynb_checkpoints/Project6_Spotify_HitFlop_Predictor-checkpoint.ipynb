### Multicollinearity

```python
# Check for multi-collinearity for Logistic Regression Model
vifs=pd.Series([VIF(df_fin.drop('target',axis=1).values,i) for i in range(df_fin.drop('target',axis=1).shape[1])],index=df_fin.drop('target',axis=1).columns)
vifs[vifs>10]
```

```python
#Function to drop multi-collinear columns and recalculating VIFS
drop_col=[]
def func(column, X):
    if column:
        X=X.drop(column, axis=1)
    vifs=pd.Series([VIF(X.values,i) for i in range(X.shape[1])],index=X.columns)
    display(vifs[vifs>10])
    drop_col.append(column)
    return X
```

```python
df_fin=func('time_signature',df_fin)
```

```python
df_fin=func('energy',df_fin)
```

```python
df_fin=func('danceability',df_fin)
```

### Dropping outliers > 95% and < 5%

```python
# 1st method - IQR Method (Not Adopted)
stats=df_fin.describe()
stats=stats.T
stats['IQR']=stats['75%']-stats['25%']

outliers = pd.DataFrame(columns=df_fin.columns)

for col in stats.index:
    iqr=stats.IQR[col]
    cutoff=iqr*1.6
    lower=stats.at[col,'25%']-cutoff
    upper=stats.at[col,'75%']+cutoff
    results=df_fin[(df_fin[col]<lower)|(df_fin[col]>upper)].copy()
    results['Outlier']=col
    outliers=outliers.append(results)
```

### Creating Dummy Variables

```python
#Create a new DF df_fin_1 with Dummy Variables 
#NOTE: I will not drop first column, since I will undergo feature engineering which doesnt require dropping it
df_fin_1=pd.get_dummies(data=df_fin, columns=dum_lst)
```

```python
X1=df_fin_1.drop('target',axis=1).reset_index(drop=True)
y1=df_fin_1['target'].reset_index(drop=True)
```

### Feature Engineering

#### Filter-based

```python
# F-Classification
selection=SelectKBest(f_classif, k=7)
selection.fit(X1,y1)
X_1_1=selection.transform(X1)
X_1_1=pd.DataFrame(X_1_1)
X_1_1
```

```python
# Chi-2
selection=SelectKBest(chi2, k=7)
selection.fit(X1,y1)
X_1_2=selection.transform(X1)
X_1_2=pd.DataFrame(X_1_2)
X_1_2
```

#### Wrappers

```python
# RFE - Logistic Regression
estimator=LogisticRegression(max_iter=1000)
selector=RFE(estimator, 7)
selector.fit(X1,y1)
X1.columns[selector.support_].to_list()
X_1_3=df_fin_1[X1.columns[selector.support_].to_list()]
X_1_3.reset_index(drop=True)
```

```python
# SFS - Logistic Regression
estimator=LogisticRegression(max_iter=1000)
s1=sfs(estimator, 7, forward=True).fit(X1,y1)
s1.get_metric_dict()
s_lst=list(s1.get_metric_dict()[7]['feature_names'])
X_1_4=df_fin_1[s_lst]
X_1_4.reset_index(drop=True)
```

```python
fig=plot_sfs(s1.get_metric_dict(), kind='std_err')
plt.title("Sequential Forward Selection")
plt.show()
```

#### PCA

```python
# PCA
pca1=PCA(0.9)
X_1_5=pd.DataFrame(pca1.fit_transform(X1))
X_1_5
```

### Modeling

```python
y1.value_counts(normalize=True)
```

#### Logistic Regression

```python
# Splitting and Assigning Test and Train for 5 optimized DF with Logistic Regression
thresholds=np.linspace(0,1,100)
lst=[X_1_1, X_1_2, X_1_3, X_1_4, X_1_5]
lst2=['FClassification', 'Chi-2', 'RFE - Logistic Regression', 'SFS - Logistic Regression', 'PCA']
count=0
dct={}

print('\nLOGISTIC REGRESSION MODEL\n')

for i in lst:
    X_train,X_test,y_train,y_test=train_test_split(i,
                                                   y1,
                                                   test_size=0.3,
                                                   random_state=42,
                                                   stratify=y1)
    
    model1=LogisticRegression(max_iter=10000, class_weight='balanced')
    model1.fit(X_train,y_train)
    y_pred={i:(model1.predict_proba(X_test)[:,1]>i).astype(int) for i in thresholds}
    
    for x,y in y_pred.items():
        conf=confusion_matrix(y_test, y)
        ac=round(accuracy_score(y_test, y),3)
        if (conf[0][1]==0) & (conf[1][1]==0):
            prec=0
        else:
            prec=round(metrics.precision_score(y_test,y),3)
        rec=round(metrics.recall_score(y_test,y),3)
        f1=round(metrics.f1_score(y_test,y),3)
        roc=round(metrics.roc_auc_score(y_test,y),3)
        dct[round(x,3)]=(ac,prec,rec,f1,roc)

    
    print('\n',lst2[count].upper(),'\n Optimal Threshold, Accuracy, Precision, Recall, F1-Score, Roc: ',max(dct.items(), key = lambda x: x[1]))
    count+=1
```

#### Decision Trees

```python
# Random Forest, Decision Tree Classifier, and XGBoost
model_list = [RandomForestClassifier, DecisionTreeClassifier, xgb.XGBClassifier]

for model in model_list:
    X_train,X_test,y_train,y_test=train_test_split(df_fin.drop('target',axis=1),
                                                   df_fin['target'],
                                                   test_size=0.3,
                                                   random_state=42,
                                                   stratify=y1)
    print(((str(model).split('.')[-1]).upper())[:-2])
    model2 = model().fit(X_train, y_train)
    y_pred2 = model2.predict(X_test)
    print('Accuracy: ', round(accuracy_score(y_test, y_pred2),3))
    print('Precision: ', round(metrics.precision_score(y_test,y_pred2),3))
    print('Recall: ', round(metrics.recall_score(y_test,y_pred2),3))
    print('F1 Score: ', round(metrics.f1_score(y_test,y_pred2),3))
    print('Roc: ', round(metrics.roc_auc_score(y_test,y_pred2),3), '\n')
    
    importances = model2.feature_importances_  
    indices = np.argsort(importances)[::-1]
    # Print the feature ranking
    print("Feature ranking:")
    for f in range(X_test.shape[1]):
        print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))
    # Plot the feature importances of the forest
    plt.figure()
    plt.title("Feature importances")
    plt.bar(range(X_test.shape[1]), importances[indices], color="r", align="center")
    plt.xticks(range(X_test.shape[1]), indices)
    plt.xlim([-1, X_test.shape[1]])
    plt.show()
```
